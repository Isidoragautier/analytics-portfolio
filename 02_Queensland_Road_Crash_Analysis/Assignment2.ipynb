```python
missing_count = crash.isnull().sum()
print("Missing values per column:\n",missing_count)
```

```python
for column in crash.columns:
    print(f"Column: {column}")
    print(crash[column].unique())
    print("-" * 50)
```

```python
crash = crash[ (crash['Crash_Longitude'].notna()) & (crash['Crash_Latitude'].notna()) ]
```

```python
missing_count = crash.isnull().sum()
print("Missing values per column:\n",missing_count)
```

```python
crash['Loc_Post_Code'] = crash['Loc_Post_Code'].astype(str)
```

```python
crash['Crash_Date'] = pd.to_datetime(crash['Crash_Year'].astype(str) + '-' + crash['Crash_Month'].astype(str) + '-01')


crash['Crash_Year_Month'] = crash['Crash_Date'].dt.to_period('M').astype(str)
crash['Crash_Year'] = crash['Crash_Date'].dt.year
crash['Crash_Month'] = crash['Crash_Date'].dt.month
```

```python
crash['Day_of_Week'] = crash['Crash_Date'].dt.day_name()

crash['Week_Type'] = crash['Day_of_Week'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
```

```python
crash.to_csv('cleaned_data.csv', index=False)
```

```python
df_sample = crash.sample(n=1000, random_state=42)
df_sample.to_csv('reduced_data.csv', index=False)
```
